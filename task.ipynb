# Visualize the Distribution of Categorical and Continuous Data

# Import Python Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("world_population.csv")
```

```python
data
```

## Sorted the Data according to their Rank and then selected the top 15 Countries

```python
data1 = data.sort_values(by='Rank', ascending=True).head(15)
```

```python
data1
```

## Plot the Bar Graph to show that which country have highest population or plot the countries according to their population 

```python
plt.figure(figsize=(14,10))
plt.bar(data1['Country/Territory'],data1['2022 Population'],color="red")
plt.xlabel('Top 15 Countries')
plt.ylabel('2022 Population')
plt.title('Counytries with highest population')
plt.show()
```

## Plot the Continents according to World population percentag.To know that which continent have highest world population percentage .

```python
plt.figure(figsize=(8,8))
plt.bar(data['Continent'],data['World Population Percentage'],color='blue')
plt.xlabel('Continent')
plt.ylabel('World Population Percentage')
plt.title('Which Continent have Highest World Population Percentage')
plt.show()
```
